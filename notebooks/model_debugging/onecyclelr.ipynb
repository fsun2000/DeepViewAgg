```python
import torch
from torch.optim.lr_scheduler import OneCycleLR
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
model = torch.nn.Linear(30, 10)
model
```

```python
max_lr = 0.1
epochs = 20
steps_per_epoch = 200
anneal_strategy = 'linear'
pct_start=0.01

optimizer = torch.optim.SGD(model.parameters(), lr=0.1, momentum=0.9)
scheduler = OneCycleLR(optimizer, max_lr=max_lr, steps_per_epoch=steps_per_epoch, epochs=epochs,
                      anneal_strategy=anneal_strategy,
                      pct_start=pct_start)

```

```python
lr_rates = []
for i in range(epochs * steps_per_epoch):
    if i % 2 == 0:
        lr_rates.append(scheduler.get_last_lr()[0])
    scheduler.step()
    
plt.plot(lr_rates)

# 0.08 - 0.1
```
